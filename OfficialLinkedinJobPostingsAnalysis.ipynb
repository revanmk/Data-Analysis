```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Job_Postings = pd.read_csv('postings.csv')
Job_Postings
```

```python
Job_Postings=Job_Postings.drop(columns=['pay_period','description','views','formatted_work_type','applies','remote_allowed','job_posting_url', 'application_url',
       'application_type','formatted_experience_level','skills_desc','posting_domain','sponsored', 'work_type', 'currency',
       'compensation_type', 'normalized_salary', 'zip_code', 'fips','closed_time','med_salary'])
```

```python
Job_Postings
```

```python
Job_Postings=Job_Postings.fillna('Unknown')
Job_Postings
```

```python
Job_Postings['max_salary']=Job_Postings['max_salary'].astype(str).replace('Unknown','0').astype(float)
Job_Postings['min_salary']=Job_Postings['min_salary'].astype(str).replace('Unknown','0').astype(float)
Job_Postings
```

```python
# benefits.csv
benefits=pd.read_csv('benefits.csv')
benefits=benefits.drop(columns=['inferred'])
benefits
```

```python
# companies.csv
companies = pd.read_csv('companies.csv')
companies = companies[['company_id','name','country']]
companies=companies.fillna('Unknown')
companies
```

```python
# employee_counts.csv
employee_counts = pd.read_csv('employee_counts.csv')
employee_counts=employee_counts.drop(columns=['follower_count'])
employee_counts=employee_counts.fillna('Unknown')
employee_counts
```

```python
# industries.csv
industries = pd.read_csv('industries.csv')
industries
```

```python
industries.industry_name.value_counts()
```

```python
# job_industries.csv
job_industries = pd.read_csv('job_industries.csv')
job_industries
```

```python
industry_jobIndustry=pd.merge(job_industries,industries,on='industry_id',how='left')
industry_jobIndustry
```

```python
# job_skills.csv
job_skills = pd.read_csv('job_skills.csv')
job_skills
```

```python
# skills.csv
skills = pd.read_csv('skills.csv')
skills
```

```python
skills_job_skills = pd.merge(skills,job_skills,on='skill_abr',how='right')
skills_job_skills
```

```python
skills_job_skills.skill_abr.value_counts().values
# y=skills_job_skills.skill_abr.value_counts().index()
```

```python
plt.figure(figsize=(12,12))
# plt.barh(y=skills_job_skills.skill_abr.value_counts().values,width=skills_job_skills.skill_abr.value_counts().index)
plt.plot(skills_job_skills.skill_abr.value_counts().values,skills_job_skills.skill_abr.value_counts().index)


# Labels and title
plt.xlabel("Number of Job Postings")
plt.ylabel("Skill Name")
plt.title("Top Skills by Job Postings")

# Show plot
plt.show()
```

```python
# salaries.csv
salaries = pd.read_csv('salaries.csv')
salaries['max_salary']=salaries['max_salary'].fillna(0)
salaries['min_salary']=salaries['min_salary'].fillna(0)
salaries['med_salary']=salaries['med_salary'].fillna(0)
salaries=salaries.drop(columns=['pay_period','currency','compensation_type'])
salaries
```
