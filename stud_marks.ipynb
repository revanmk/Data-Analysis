```python
daDf=data.iloc[:,[0]+list(range(56,107))]
daDf
```

```python
NewDf=daDf.copy()
NewDf.drop(['CANDIDATE ID','TOTR'],axis=1,inplace=True)
NewDf
```

```python
Marks=[]
Right=[]
Wrong=[]
n=0
while(n<NewDf.shape[0]):
    marks=0
    right=0
    wrong=0
    for i in NewDf:
        if NewDf[i][n]==NewDf[i][0] or i=='Q70':
            marks+=1
            right+=1
        else:
            wrong+=1
    Marks.append(marks)
    Right.append(right)
    Wrong.append(wrong)
    n+=1
Marks

            
```

```python
daDf['Da Marks']=Marks
daDf['Right DA']=Right
daDf['Wrong DA']=Wrong
daDf
```

```python
daDf.columns = daDf.columns.str.strip()
daDf
```

```python
daDf.sort_values(by='Da Marks', ascending=False, inplace=True)
daDf
```

```python
daDf['Rank'] = list(range(0,daDf.shape[0]))
daDf
```

```python
RankDf=daDf[['Rank','CANDIDATE ID','Da Marks','TOTR']]
RankDf['DSA MArks']=RankDf['TOTR']-RankDf['Da Marks']
RankDf
```
